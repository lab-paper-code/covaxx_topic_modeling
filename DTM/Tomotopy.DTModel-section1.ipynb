```python
import tomotopy as tp
import numpy as np
import nltk
from tqdm import tqdm

```

```python
def data_feeder(input_file):
    for line in open(input_file, encoding='utf-8'):
        fd = line.strip().split(maxsplit=1)
        #print(line)
        timepoint = int(fd[0])
        #print(timepoint)
        yield fd[1], None, {'timepoint':timepoint}
```

```python
porter_stemmer = nltk.PorterStemmer().stem
corpus = tp.utils.Corpus(
    tokenizer=tp.utils.SimpleTokenizer(porter_stemmer)
)

corpus.process(data_feeder("./Datasets/dtm_input_section1.txt"))
```

```python
#n-gram
# cands = corpus.extract_ngrams(min_cf=20, min_df=10, max_len=3, max_cand=1000, normalized=False)
# corpus.concat_ngrams(cands, delimiter='_')
coherence = []
for num in tqdm(range(10,55,5)):
    
    mdl = tp.DTModel(min_cf=3, k=num, t=4, corpus=corpus)
    mdl.train(0)


    #print('Num docs:{}, Num Vocabs:{}, Total Words:{}'.format(
    #    len(mdl.docs), len(mdl.used_vocabs), mdl.num_words
    #))
    #print('Removed Top words: ', *mdl.removed_top_words)

    for i in range(0, 5000, 20):
        print('Iteration: {:04}, LL per word: {:.4}'.format(i, mdl.ll_per_word))
        mdl.train(20)
    #print('Iteration: {:04}, LL per word: {:.4}'.format(5000, mdl.ll_per_word) + "\n")

    #print(mdl.summary())
    mdl.save(f'./models/phase1_dtm_number_of_topics_{num}.bin')
    coh = tp.coherence.Coherence(mdl, coherence='u_mass')
    coherence.append(coh.get_score())
    
    '''topic_dist_by_time = np.zeros(shape=[mdl.num_timepoints, mdl.k], dtype=np.float64)
    for doc in mdl.docs:
        topic_dist_by_time[doc.timepoint] += doc.get_topic_dist()

    topic_dist_by_time /= mdl.num_docs_by_timepoint[:, np.newaxis]

    for k in range(mdl.k):
        print('Topic #{}'.format(k), *(w for w, _ in mdl.get_topic_words(k, 0, top_n=50)))
        print(topic_dist_by_time[:, k])

    for k in range(mdl.k):
        for t in range(mdl.num_timepoints) :
            mdl.get_topic_words(k, t, top_n = 50)'''

print('end')
```

```python
def get_topic_diversity(timepoint):
    unique_words = set([])
    for k in range(dtm_model.k):
        temp_list = [word[0] for word in dtm_model.get_topic_words(timepoint=timepoint, topic_id =k)]
        unique_words = unique_words.union(set(temp_list))
    return len(unique_words)/(10*dtm_model.k)
```

```python
for num in range(10,55,5):
    dtm_model = dtm_model = tp.DTModel().load(f'./models/section1_model.bin')
    td = 0
    for timepoint in range(dtm_model.num_timepoints):
        td += get_topic_diversity(timepoint)    
    print(num," ",td/dtm_model.num_timepoints)
```
